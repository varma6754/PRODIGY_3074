```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Define file paths
training_file = "C:\\prodigy\\twitter_training.csv"
validation_file = "C:\\prodigy\\twitter_validation.csv"
```

```python
headers = ["ID", "Category", "Sentiment", "Tweet"]
```

```python
training_data = pd.read_csv(training_file, names=headers, encoding='ISO-8859-1')
validation_data = pd.read_csv(validation_file, names=headers, encoding='ISO-8859-1')
```

```python
combined_data = pd.concat([training_data, validation_data], ignore_index=True)
```

```python
training_data.head()
```

```python
validation_data.head()
```

```python
combined_data.dropna(inplace=True)
```

```python
print(combined_data.head())
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=combined_data, x='Sentiment', order=combined_data['Sentiment'].value_counts().index, palette='viridis')
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(data=combined_data, x='Category', hue='Sentiment', palette='viridis')
plt.xticks(rotation=90)
plt.title('Sentiment Distribution by Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.legend(title='Sentiment')
plt.show()
```

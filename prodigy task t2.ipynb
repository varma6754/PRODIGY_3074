```python
fig, axes = plt.subplots(1, 3, figsize=(20, 5))

# Plot 1: Survival rate by passenger class with custom colors
sns.barplot(x='Pclass', y='Survived', data=train_df, ax=axes[0], palette='Set2')
axes[0].set_title('Survival Rate by Passenger Class')

# Plot 2: Survival rate by gender with different color hues
sns.barplot(x='Sex', y='Survived', data=train_df, ax=axes[1], hue='Pclass', palette='Set1')
axes[1].set_title('Survival Rate by Gender (Class Differentiated)')

# Plot 3: Survival rate by embarkation port with custom colors
sns.barplot(x='Embarked', y='Survived', data=train_df, ax=axes[2], palette='husl')
axes[2].set_title('Survival Rate by Embarkation Port')

# Show plots
fig.tight_layout()
plt.show()
```
